```python
import numpy as np
from cosmoTransitions import generic_potential as gp
from cosmoTransitions import pathDeformation as pd
```

```python
class model(gp.generic_potential):
    def init(self, lh, A, muH, muS):
        self.muH = muH
        self.lh = lh
        self.muS = muS
        self.A = A
        self.eps = 0.001

        self.Ndim = 2
        self.renormScaleSq = 150.0**2

        self.g = 0.65
        self.gY = 0.36
        self.yt = 0.9945


    def V0(self,X):
        X = np.asanyarray(X)
        phi1,phi2 = X[...,0],X[...,1]
        r = -0.5*(self.muH**2)*(phi1**2) + 0.25*self.lh*phi1**4 + 0.5*(self.muS**2)*(phi2**2) - 0.5*self.A * phi2 * phi1**2
        return r
    
    def boson_massSq(self,X):
        X = np.asanyarray(X)
        phi1,phi2 = X[...,0],X[...,1]

        sqrt = self.A**2 * (4*phi1**2 + phi2**2)+2*self.A*phi2*(self.muH**2 + self.muS**2 - 3*self.lh*phi1**2)+(self.muH**2 + self.muS**2 - 3*self.lh*phi1**2)**2

        mhsq = 0.5*(-self.muH**2 + self.muS**2 - self.A*phi2 + 3*self.lh*phi1**2 + np.sqrt(sqrt))
        mSsq = 0.5*(-self.muH**2 + self.muS**2 - self.A*phi2 + 3*self.lh*phi1**2 - np.sqrt(sqrt))

        mXsq = -self.muH**2 -self.A*phi2 + self.lh*phi1**2

        mWsq = 0.25*self.g**2 * phi1**2
        mZsq = 0.25*(self.g**2 + self.gY**2)*phi1**2

        M = np.array([mhsq,mSsq,mXsq,mWsq,mZsq])

        M = np.rollaxis(M,0, len(M.shape))

        dof = np.array([1,1,3,6,3])

        c = np.array([1.5,1.5,1.5,5./6., 5./6.])

        return M, dof, c

    def fermion_massSq(self,X):
        X = np.asanyarray(X)
        phi1 = X[...,0]

        mt = 0.5*self.yt**2. * phi1**2.
        M = np.array([mt])

        M = np.rollaxis(M,0,len(M.shape))

        dof = np.array([12])

        return M, dof

    def approxZeroTMin(self):
        return [np.array([246., 600.])]

    def VCW(self, bosons, fermions):
        m2, n, c = bosons
        y = np.sum(n*m2*m2 * (np.log(m2/self.renormScaleSq + 1e-100 + 0j)
                              - c), axis=-1)
        m2, n = fermions
        c = 1.5
        y -= np.sum(n*m2*m2 * (np.log(m2/self.renormScaleSq + 1e-100 + 0j)
                               - c), axis=-1)
        return y.real/(64*np.pi*np.pi)

    def V0T(self, X):
        X = np.asanyarray(X, dtype=float)
        bosons = self.boson_massSq(X)
        fermions = self.fermion_massSq(X)

        V0T = self.V0(X)
        V0T += self.VCW(bosons,fermions)

        return V0T

    def dV(self, X):
        "Ridders algorithm"
        X = np.array(X, dtype=float)
        eps = self.eps
        dVdx = (self.V0T(np.array([X[...,0]-2.*eps,X[...,1]]))-8.*self.V0T(np.array([X[...,0]-eps,X[...,1]]))+8.*self.V0T(np.array([X[...,0]+eps,X[...,1]]))-self.V0T(np.array([X[...,0]+2.*eps,X[...,1]])))/(12.*eps)
        dVdy = (self.V0T(np.array([X[...,0],X[...,1]-2.*eps]))-8.*self.V0T(np.array([X[...,0],X[...,1]-eps]))+8.*self.V0T(np.array([X[...,0],X[...,1]+eps]))-self.V0T(np.array([X[...,0],X[...,1]+2.*eps])))/(12.*eps)
        rval = np.empty_like(X)
        rval[...,0], rval[...,1] = dVdx, dVdy
        return rval

    def plot2d(self, box, treelevel=False,offset=0, xaxis=0, yaxis=1, n=50, clevs=200,cfrac=.8, **contourParams):
        import matplotlib.pyplot as plt
        xmin, xmax, ymin, ymax = box
        X = np.linspace(xmin,xmax,n).reshape(n,1)*np.ones((1,n))
        Y = np.linspace(ymin,ymax,n).reshape(1,n)*np.ones((n,1))
        XY = np.zeros((n,n,self.Ndim))
        XY[...,xaxis],XY[...,yaxis] = X,Y
        XY += offset
        Z = self.V0(XY) if treelevel else self.V0T(XY)
        minZ,maxZ = min(Z.ravel()),max(Z.ravel())
        N = np.linspace(minZ,minZ+(maxZ - minZ)*cfrac, clevs)
        plt.contour(X,Y,Z, N, **contourParams)
        plt.axis(box)
        plt.show()

    def plot1d(self, xmin, xmax, treelevel=False, offset=0,n=50):
        "This function makes the singlet satisfy the vev condition as always"
        import matplotlib.pyplot as plt
        X = np.linspace(xmin,xmax,n)
        Y = 0.5*self.A*(X**2)/(self.muS**2)
        XY=np.zeros((n,self.Ndim))
        XY[...,0],XY[...,1] = X,Y
        XY += offset
        Z = self.V0(XY) if treelevel else self.V0T(XY)
        minZ, maxZ = min(Z.ravel()),max(Z.ravel())
        plt.plot(X,Z)
        plt.show()
```

```python
m = model(0.125299, 10.6204, 20.2598, 21.8138)
```

```python
m.plot2d(box=[0,400.,0,2e3],treelevel=True,n=1000)
```

```python
pd.fullTunneling([[500., 2789.9],[87.,83.]], m.V0T, m.dV)
```
